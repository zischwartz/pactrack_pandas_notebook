```python
# grouping by both we get the largest links between two unique pacs, the sum of the transactions between them
# this is basically equivalent to the data output in process.ipynb and use for the visualization
pac_to_pac_nets  = pac_to_pac_transactions.groupby(['CMTE_ID', 'OTHER_ID']) 
# and then sum that
pac_to_pac_nets = pac_to_pac_nets['TRANSACTION_AMT'].sum()
# and now reset the index to get rid of the group by and just make it flat
pac_to_pac_nets = pac_to_pac_nets.reset_index()
```

```python
pac_to_pac_nets.merge(pacs, on="CMTE_ID").sort_values('TRANSACTION_AMT', ascending=False).reset_index()
```

```python
# now, grouping the original df just by cmte_id means we get the total outlay of one pac to all other pacs
pac_to_pac_totals  = pac_to_pac_transactions.groupby(['CMTE_ID']) 
# and then sum that
pac_to_pac_totals = pac_to_pac_totals['TRANSACTION_AMT'].sum().reset_index()
# and now reset the index to get rid of the group by and just make it flat
pac_to_pac_totals = pac_to_pac_totals.merge(pacs, on="CMTE_ID").sort_values('TRANSACTION_AMT', ascending=False).reset_index()
pac_to_pac_totals
```
